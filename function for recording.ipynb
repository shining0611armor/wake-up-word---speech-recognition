```python
import sounddevice as sd
import numpy as np
import scipy.io.wavfile as wav

def record_audio(duration, filename):
    fs = 16000  
    print(f"Recording for {duration} seconds...")
    recording = sd.rec(int(duration * fs), samplerate=fs, channels=1, dtype='int16')
    sd.wait()  
    wav.write(filename, fs, recording)
    print(f"Recording saved to {filename}")

def record_wake_word(samples=100):
    """
    Record wake word samples.

    :param samples: Number of samples to record.
    """
    for i in range(samples):
        filename = f'wake_word_{i+1}.wav'
        record_audio(2, filename)

def record_background_sound(samples=100):
    """
    Record background sound samples.

    :param samples: Number of samples to record.
    """
    for i in range(samples):
        filename = f'background_sound_{i+1}.wav'
        record_audio(2, filename)

record_wake_word(100)
```
